```python
#Libraries to import
import nltk 
import pandas as pd
import nltk, string
import matplotlib.pyplot as plt

#Downloading gutemberg corpus
nltk.download('gutenberg')
nltk.download('stopwords')
```

### 25 non-stopwords with more number of occurrences

```python
"""
Pipeline process to get the 25 non-stopwords with more occurrences
    1. Get and tokenize the corpus
    2. Convert to lower case
    3. Get a list of stopwords from nltk
    4. Add punctuation to the list of stopwords
    5. Add other punctuation to the list of stopwords
    6. Remove stopwords from the corpus
    7. Get the frequency distribution of the words
    8. Get the 25 most common words
    9. Plot the 25 most common words
"""
#1. Get and tokenize the corpus(read corpus from nltk gutenberg)
corpus = nltk.corpus.gutenberg.words('blake-poems.txt')
#2. Converting corpus to lower case
corpus = [word.lower() for word in corpus]

#3. Get a list of stopwords from nltk
stopwords = nltk.corpus.stopwords.words('english')
#4. Add punctuation to the list of stopwords
stopwords += string.punctuation
#5. Add other punctuation to the list of stopwords (It was added more 
# punctuation to the stopwords list because the corpus has some 
# punctuation that is not in the string.punctuation list)
stopwords += ['."',]
#6.Remove stopwords from corpus
nonStopwords = [word for word in corpus if not word in stopwords]
#7. Get the frequency distribution
frequencies = {word:corpus.count(word) for word in nonStopwords}
#8. Get the 25 most common words
mostCommon = {key: val for key, val in sorted(frequencies.items(), key=lambda x:x[1], reverse=True)[:25]}

plt.figure(figsize=(14,6))
#9.plot the color more green if have more occurrences and more red if have less occurrences
plt.imshow(pd.DataFrame(mostCommon, index=[0]), cmap='Greens')
#Add the values to the plot
for i in range(len(mostCommon)):
    plt.text(i, 0, list(mostCommon.values())[i], fontsize=12, color='green', ha='center', va='center')
plt.xticks(range(len(mostCommon)), list(mostCommon.keys()), rotation=90)
plt.yticks([])
plt.show()
```

```python
#create a histogram with the 25 most common words
#plot two plot one over the other
plt.figure(figsize=(14,6))
plt.bar(mostCommon.keys(), mostCommon.values(), edgecolor='black', color='g')
plt.xticks(rotation=90)
plt.title('25 most common Non-Stopwords')
plt.xlabel('Non-stopwords')
plt.ylabel('Frequency')
plt.show()
```

```python
#Get frequency distribution(Using nltk)
fdist = nltk.FreqDist(nonStopwords)
fdist.most_common(25)
```
