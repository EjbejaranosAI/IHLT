### Importing libraries

```python
import spacy
import os
import nltk
import neuralcoref

nlp = spacy.load('en_core_web_sm')

path = "/content/drive/MyDrive/Colab_Notebooks/2.IHLT/Lab9/"
filename = '11-0.txt'

nltk.download("punkt")
```

```python
from nltk.corpus.reader.plaintext import PlaintextCorpusReader
corpus = PlaintextCorpusReader(path, filename)
print(corpus.fileids())
paragraph = corpus.paras()
print("Total Paragraphs:", len(paragraph))
```

```python
first_paragraph = paragraph[16]
print(first_paragraph)
```

```python
neuralcoref.add_to_pipe(nlp)
```

```python
doc = nlp(' '.join(first_paragraph[0]))
```

```python
doc._.has_coref
```

```python
doc._.coref_clusters
```

Conclusions

Observing the obtained results, first we notice that the main coreference clusters are found, but not all the references. The most evident is "book", for which the method found its references ("it") but not the main word.
On the other hand, this results does not match with the obtained using the Hugging Face's version of NeuralCoref, which was able to found additional clusters and references such as "pictures or conversations" (see link).
